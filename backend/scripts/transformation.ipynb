```python
def map_location_to_country(location):
    country = location.split('-')[0].strip().capitalize() 
    return country if country in country_mapping else None
```

```python
df['Location'].unique()
```

```python
transformed_df['countryId'] = df['Location'].apply(map_location_to_country).map(country_mapping)
transformed_df['countryId'].unique()
```

```python
def convert_date(date_str):
    if pd.isna(date_str):
        return None
    try:
        date_str = date_str.replace('-', ' ')
        parts = date_str.split()
        if len(parts[0]) == 1:
            parts[0] = f"0{parts[0]}"
        date_str = " ".join(parts)
        return pd.to_datetime(date_str, format='%d %b %Y').strftime('%Y-%m-%dT%H:%M:%S.%fZ')
    except:
        try:
            return pd.to_datetime(date_str).strftime('%Y-%m-%dT%H:%M:%S.%fZ')
        except:
            return None
```

```python
transformed_df['dateOfJoining'] = df['Start Date'].apply(convert_date)
```
